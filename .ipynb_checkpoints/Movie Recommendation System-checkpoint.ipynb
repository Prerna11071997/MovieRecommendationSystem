### Creating movie Recommendation

```python
df.head()
```

```python
moviemat = df.pivot_table(index = 'user_id', columns = 'title', values = 'rating' )
```

```python
moviemat
```

```python
ratings.sort_values('no of ratings', ascending = False).head()
```

```python
starwar_user_ratings = moviemat['Star Wars (1977)']
```

```python
starwar_user_ratings
```

```python
similar_to_starwars = moviemat.corrwith(starwar_user_ratings)
```

```python
corr_starwar = pd.DataFrame(similar_to_starwars,columns = ['Correlation'])
```

```python
corr_starwar.dropna(inplace = True)
```

```python
corr_starwar.head()
```

```python
corr_starwar.sort_values('Correlation', ascending = False).head(10)
```

```python
corr_starwar = corr_starwar.join(ratings['no of ratings'])
corr_starwar.head()
```

```python
corr_starwar[corr_starwar['no of ratings'] > 100].sort_values('Correlation', ascending = False)
```

### Predict Function

```python
def predict_movie(movie_name):
    movie_user_ratings = moviemat[movie_name]
    similar_to_movie = moviemat.corrwith(movie_user_ratings)
    corr_movie = pd.DataFrame(similar_to_movie ,columns = ['Correlation'])
    corr_movie.dropna(inplace = True)
    corr_movie = corr_movie.join(ratings['no of ratings'])
    predictions =  corr_movie[corr_movie['no of ratings'] > 100].sort_values('Correlation', ascending = False)
    
    return predictions
```

```python
predictions = predict_movie("Titanic (1997)")
```
